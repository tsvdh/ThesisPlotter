```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_lims = [[0, 0.1], [0, 1], [0, 5]]
y_lims = [[0, 1], [0, 1], [0, 1]]

for i in range(3):
    a = np.full(100, x_lims[i][1])

    x = np.linspace(1e-6, x_lims[i][1], 100)
    light_true = 1 / np.exp(x)
    light_reverse = 1 / np.exp(-x + a)
    light_approx = (1 - np.exp(-a)) / a

    fig = plt.figure()
    plt.xlabel("Distance in Node (x)")
    plt.ylabel("Radiance Transmittance")
    plt.plot(x, light_true, label="Rad")
    # plt.plot(x, light_reverse, label="ReverseRad")
    plt.plot(x, light_approx, label="ApproxRad")
    plt.legend()
    plt.xlim(x_lims[i])
    plt.ylim(y_lims[i])
    plt.savefig(f"transmittance_{x_lims[i][1]}.png", bbox_inches="tight")
    plt.savefig(f"transmittance_{x_lims[i][1]}.pdf", bbox_inches="tight")
    # plt.savefig(f"transmittance_{x_lims[i][1]}_reverse.png", bbox_inches="tight")
    # plt.savefig(f"transmittance_{x_lims[i][1]}_reverse.pdf", bbox_inches="tight")
```

```python
x_lims = [[0, 0.1], [0, 1], [0, 5]]
y_lims = [[0, 1], [0, 1], [0, 1]]

for i in range(3):
    a = np.full(100, x_lims[i][1])

    x = np.linspace(1e-6, x_lims[i][1], 100)
    light_true = (1 / np.exp(x)) *  (1 / np.exp(x))
    light_reverse = (1 / np.exp(x)) * (1 / np.exp(-x + a))
    light_approx = (1 / np.exp(x)) * (1 - np.exp(-a)) / a

    fig = plt.figure()
    plt.xlabel("Distance in Node(x)")
    plt.ylabel("Captured Radiance")
    plt.plot(x, light_true, label="RadCaptured")
    # plt.plot(x, light_reverse, label="ReverseRadCaptured")
    plt.plot(x, light_approx, label="ApproxRadCaptured")
    plt.legend()
    plt.xlim(x_lims[i])
    plt.ylim(y_lims[i])
    plt.savefig(f"captured_{x_lims[i][1]}.png", bbox_inches="tight")
    plt.savefig(f"captured_{x_lims[i][1]}.pdf", bbox_inches="tight")
    # plt.savefig(f"captured_{x_lims[i][1]}_reverse.png", bbox_inches="tight")
    # plt.savefig(f"captured_{x_lims[i][1]}_reverse.pdf", bbox_inches="tight")
```

```python
x_lims = [[0, 0.1], [0, 1], [0, 5]]
y_lims = [[0, 0.6], [0, 0.6], [0, 0.6]]

for i in range(3):
    x = np.linspace(1e-6, x_lims[i][1], 100)
    integral_true = 1/2 - np.exp(-2 * x) / 2
    integral_reverse = np.exp(-x) * x
    integral_approx = (1 - np.exp(-x)) / x - ((1 - np.exp(-x)) * np.exp(-x)) / x

    fig = plt.figure()
    plt.xlabel("Node Diameter (d)")
    plt.ylabel("Total Radiance Captured")
    plt.plot(x, integral_true, label="RadCapturedTotal")
    # plt.plot(x, integral_reverse, label="ReverseRadCapturedTotal")
    plt.plot(x, integral_approx, label="ApproxRadCapturedTotal")
    plt.legend()
    plt.xlim(x_lims[i])
    plt.ylim(y_lims[i])
    plt.savefig(f"total_{x_lims[i][1]}.png", bbox_inches="tight")
    plt.savefig(f"total_{x_lims[i][1]}.pdf", bbox_inches="tight")
    # plt.savefig(f"total_{x_lims[i][1]}_reverse.png", bbox_inches="tight")
    # plt.savefig(f"total_{x_lims[i][1]}_reverse.pdf", bbox_inches="tight")
```
