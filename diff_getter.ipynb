```python
import numpy as np

default_path = "C:/Users/tsvdh/OneDrive/Documenten/thesis/results/cube_sizes_subsection/volpath_spp/mse_sppSPP.txt"

samples = [16, 32, 64, 128, 256, 512, 1024]

absolute_error = np.zeros(len(samples))
positive_error = np.zeros(len(samples))
negative_error = np.zeros(len(samples))

rounding = 5

for i, spp in enumerate(samples):
    path = default_path.replace("SPP", str(spp))

    with open(path) as file:
        absolute_error[i] = round(float(file.readline()), rounding)
        positive_error[i] = round(float(file.readline()), rounding)
        negative_error[i] = round(float(file.readline()), rounding)

negative_error = np.absolute(negative_error)

print(absolute_error)
print(positive_error)
print(negative_error)
```
